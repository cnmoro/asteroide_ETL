```python
for i in df.columns:
    new_col = f"{i}_binned"
    result, bins = pd.qcut(df[i], 5, retbins=True, labels=False) # limite da direita é aberto, limite da esquerda é fechado
    print(f'Bins {i}: {bins}')                                   # exemplo: 0.5 <= val < 10
    df[new_col] = pd.qcut(df[i], 5, labels=bins[:-1])


# df.drop(['Sales_Bins'],inplace=True,axis=1)
# result, bins=pd.cut(df.Sales,retbins=True,bins = 4)
# print(bins)
```
